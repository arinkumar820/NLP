# üîç **What is Semantic Search Anyway?**

Great question! ü§ì

Instead of scanning for exact keywords, semantic search looks at **what you mean**.
It turns each paragraph (and your question) into **embeddings** ‚Äî fancy vectors that represent meaning in mathematical form. üìä

The system then compares your query vector with all paragraph vectors to find the *closest match*.
That‚Äôs how it pulls only the pages that actually matter ‚Äî so the model focuses on just the good stuff. ‚ú®

* You upload a PDF ‚Üí stored in cloud storage (like AWS S3 ‚òÅÔ∏è).

# üîß LangChain Components Overview

**LangChain** is an open-source framework built to simplify the development of powerful applications using **Large Language Models (LLMs)**. It takes care of the complexity involved in orchestrating multiple components like models, memory, embeddings, and tools ‚Äî making it easier to build apps like chatbots, document assistants, and autonomous agents.

One of LangChain‚Äôs biggest advantages is that it's **model-agnostic** ‚Äî meaning you can easily switch between LLM providers like OpenAI, Anthropic, or Google Gemini without rewriting everything.

LangChain organizes the LLM application landscape into six main components:

> **Models, Prompts, Chains, Memory, Indexes, and Agents**

---

## ü§ñ 1. Models

The **Models** component handles interaction with the actual AI models. Since different providers (OpenAI, Hugging Face, Mistral AI, etc.) have different APIs, LangChain offers a **standardized interface** for working with all of them.

There are two types of models:

* **Language Models**: Input text ‚Üí output text. Used in chatbots, assistants, summarizers, etc.
* **Embedding Models**: Input text ‚Üí output vector. Used in semantic search and retrieval.

LangChain supports many popular model providers, allowing you to easily switch providers or plug in open-source models.

---

## üß† 2. Prompts

**Prompts** are the messages you send to LLMs ‚Äî and how you craft them makes a *huge* difference. LangChain gives you full control and flexibility over prompt design.

Key types of prompts include:

* **Dynamic Prompts**: Templates with placeholders that are filled at runtime based on user input.
* **Role-Based Prompts**: Instruct the model to behave like a specific persona (e.g., ‚ÄúYou are a medical expert‚Ä¶‚Äù).
* **Few-Shot Prompts**: Provide examples to guide the model‚Äôs response style or logic.

> The art of writing effective prompts is known as **Prompt Engineering** ‚Äî and LangChain makes it easy to experiment.

---

## üîó 3. Chains

**Chains** allow you to build powerful workflows by connecting multiple steps together. Instead of manually handling input-output between steps, LangChain handles it for you.

Types of chains include:

* **Sequential Chains**: Run one step after another.
* **Parallel Chains**: Run multiple steps at the same time and combine results.
* **Conditional Chains**: Dynamically choose the next step based on logic or prior output.

Chains are what turn isolated model calls into full applications.

---

## üìö 4. Indexes

**Indexes** allow your LangChain app to access external documents like PDFs, websites, or private databases ‚Äî so the model can answer questions based on real-world or custom data.

The indexing pipeline typically includes:

1. **Document Loader**: Brings in documents from local or cloud sources (e.g., Google Drive, web URLs).
2. **Text Splitter**: Breaks large documents into smaller chunks.
3. **Embedding Model**: Converts each chunk into a vector.
4. **Vector Store**: Stores vectors in a searchable database.
5. **Retriever**: Finds the most relevant chunks based on a user query.

This setup powers **RAG (Retrieval-Augmented Generation)** ‚Äî where the model is grounded in your data, not just its training.

---

## üí¨ 5. Memory

LLMs are stateless ‚Äî meaning they forget previous messages unless you include the full conversation every time. The **Memory** component fixes this by giving your app the ability to remember what was said before.

Popular types of memory:

* **Conversation Buffer Memory**: Remembers everything in the session.
* **Buffer Window Memory**: Keeps only the last *N* messages.
* **Summarizer Memory**: Stores a summary instead of full chat history to reduce token usage.
* **Custom Memory**: Saves user-specific data like preferences or known facts.

This is essential for building **conversational experiences** that feel natural and intelligent.

---

## üß†üõ†Ô∏è 6. Agents

**Agents** are where LangChain goes beyond basic chat ‚Äî into real-world action. While a chatbot only answers questions, **an AI Agent reasons, plans, and uses tools**.

Think of an agent as an AI that can *do things*, like:

* Search the web üåê
* Query a database üìä
* Use a calculator üßÆ
* Book a flight ‚úàÔ∏è

### üß† Agent Reasoning Example:

You ask: *‚ÄúMultiply today‚Äôs temperature in Delhi by 3.‚Äù*

The agent:

1. Recognizes it needs the temperature.
2. Uses a **Weather API** to get it.
3. Uses a **Calculator Tool** to multiply.
4. Responds with the answer.

It breaks the problem into steps using logic ‚Äî often with **Chain-of-Thought** reasoning.

| ----------------- | -------------------- | ------------------------------------ |
| **Core Function** | Understand & respond | Understand, respond, **and act**     |
| **Reasoning**     | Limited              | ‚úÖ Step-by-step logic                 |
| **Tool Usage**    | ‚ùå Not supported      | ‚úÖ Access to APIs, tools, files       |
| **Experience**    | Conversational only  | Conversational + **Action-Oriented** |

## LangChain Components Overview
- LangChain is an open-source framework used to build applications powered by Large Language Models (LLMs).   
- It helps manage the complexity of creating LLM applications by providing orchestration and pipeline building capabilities.    
- A key benefit is its **model-agnostic** nature, allowing easy switching between different LLM providers (like OpenAI, Google Gemini) with minimal code changes.    
- LangChain applications include conversational chatbots, AI knowledge assistants, and agents.   
- The framework is structured around six main components: **Models**, **Prompts**, **Chains**, **Memory**, **Indexes**, and **Agents**.    

---

## Models Component
- **Models** are the core interface for interacting with AI models in LangChain.    
- They solve the problem of different LLM providers having varying API implementations by providing a standardized way to communicate with any AI model.      
- LangChain supports two main types of models:    
    -   **Language Models**: These are typically LLMs that take text as input and return text as output.     They are used for applications like chatbots and agents.   
    -   **Embedding Models**: These models take text as input and return a vector representation as output.    They are primarily used for semantic search.   
- LangChain provides interfaces to various model providers for both Language Models (e.g., OpenAI, Anthropic, Mistral AI, Hugging Face) and Embedding Models (e.g., OpenAI, Mistral AI, IBM).       

---

## Prompts Component
- **Prompts** are the inputs provided to an LLM.    
- The output of an LLM is highly dependent on the prompt; even small changes can significantly alter the response.    
- The field of **Prompt Engineering** has emerged focusing on designing effective prompts.    
- LangChain offers flexibility in creating different types of prompts:   
    -   **Dynamic and Reusable Prompts**: Using placeholders that can be filled later based on user input, allowing the same prompt structure to be used for different topics or tones.     
    -   **Role-Based Prompts**: Defining a system-level prompt to instruct the LLM to act as a specific persona (e.g., an experienced doctor) before providing the user's query.      
    -   **Few-Shot Prompts**: Providing the LLM with a few examples of input-output pairs to guide its response format or task performance before asking the final question.         

---

## Chains Component
- **Chains** are a crucial component used to build pipelines in LangChain.     
- They automate the flow of data between different stages of an application, automatically taking the output of one stage and using it as the input for the next, eliminating the need for manual coding of this process.      
- Chains can represent complex workflows beyond simple sequential steps.   
- Examples of chain types include:
    -   **Sequential Chains**: Stages execute one after another.   
    -   **Parallel Chains**: Multiple processes execute simultaneously on the same input, and their results can be combined later.      
    -   **Conditional Chains**: The execution path depends on a condition or the result of a previous step.     

---

## Indexes Component
- **Indexes** connect your LangChain application to external knowledge sources like PDFs, websites, or databases.     
- They are used to build applications that can answer questions based on private or specific data that the LLM wasn't trained on.      
- Indexes are composed of four main parts:   
    1.  **Document Loader**: Loads data from various sources (e.g., Google Drive, databases).    
    2.  **Text Splitter**: Breaks down large documents into smaller, manageable chunks for processing.     
    3.  **Vector Store**: Stores the embedding vectors generated from the text chunks.      This is often a specialized database called a vector database.   
    4.  **Retriever**: Takes a user query, generates its embedding, performs a semantic search in the vector store to find relevant chunks, and retrieves them.     
- The retrieved relevant information is then provided to the LLM along with the user's query to generate an informed response.

## Memory Component
- **Memory** addresses the problem that standard LLM API calls are **stateless**, meaning they do not remember previous interactions in a conversation.     
- This statelessness makes building conversational applications like chatbots frustrating as the model forgets the context of the discussion.   
- The Memory component adds the ability for the application to remember the history of a conversation.   
- Common types of memory in LangChain include:
    -   **Conversation Buffer Memory**: Stores the entire history of the conversation.    
    -   **Conversation Buffer Window Memory**: Stores only the last 'N' interactions in the conversation, limiting the history size.   
    -   **Summarizer Based Memory**: Stores a summary of the conversation history instead of the full text, saving processing cost.   
    -   **Custom Memory**: Used for advanced cases to store specific, persistent information about the user, such as preferences or facts.

## Types of Models
- LangChain primarily supports two types of models: **Language Models** and **Embedding Models**.   
- **Language Models** take text as input, process it, and return text as output. They are used for applications like chatbots.     
- **Embedding Models** take text as input and return a series of numbers (vectors) representing the contextual understanding of the text. These vectors are called **embeddings**.     

## Language Models: LLMs vs. Chat Models
- Language Models are further divided into **LLMs** (Large Language Models) and **Chat Models**.   
- **LLMs** are general-purpose models for various NLP tasks like text generation, summarization, code generation, and question answering. They take a plain text string and return a plain text string.    
- **Chat Models** are specialized for conversational tasks and take a sequence of messages as input, returning chat messages as output. They can understand the entire conversation history.     
- **Important:** Chat Models are increasingly preferred for building AI applications like chatbots and agents, and LangChain's recent versions encourage their use over LLMs.      

| Feature         | LLMs (Large Language Models)          | Chat Models                           |
| :-------------- | :------------------------------------ | :------------------------------------ |
| **Purpose**     | General-purpose NLP tasks             | Specialized for Conversation Tasks    |
| **Input**       | Plain text string                     | Sequence of messages                  |
| **Output**      | Plain text string                     | Sequence of chat messages             |
| **LangChain Support** | Decreasing, not recommended for new projects    | Increasing, recommended for new projects    |
       

## Using Language Models (LLMs)
- You can use LLMs by creating an object of the chosen model class (e.g., `OpenAI`) and calling the `.invoke()` method with your prompt string.     
- The `.invoke()` method sends the prompt to the model, processes the response, and returns it.    
- Example using OpenAI's `gpt-3.5-turbo-instruct` model:
  ```python
  from langchain_openai import OpenAI
  from dotenv import load_dotenv

  load_dotenv() # Load API key from .env file

  # Create LLM object
  llm = OpenAI(model="gpt-3.5-turbo-instruct")

  # Invoke the model with a prompt
  result = llm.invoke("What is the capital of India?")

  # Print the result
  print(result)
  ```
              

## Using Chat Models
- Chat models are used similarly but take messages as input/output.
- Example using OpenAI's `gpt-4` model:
  ```python
  from langchain_openai import ChatOpenAI
  from langchain_core.messages import HumanMessage
  from dotenv import load_dotenv

  load_dotenv()

  # Create Chat Model object
  chat_model = ChatOpenAI(model="gpt-4")

  # Invoke the model with a message list
  result = chat_model.invoke([HumanMessage(content="What is the capital of India?")])

  # Print the content of the result
  print(result.content)
  ```
      
- You can also use open-source chat models via Hugging Face's Inference API.    
- This requires a Hugging Face account and an access token stored in your `.env` file as `HF_HUB_ACCESS_TOKEN`.     
- Example using Hugging Face API with a `tinyllama` model:
  ```python
  from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
  from langchain_core.messages import HumanMessage
  from dotenv import load_dotenv
  import os

  load_dotenv()

  # Get Hugging Face token from environment
  hf_token = os.environ.get("HF_HUB_ACCESS_TOKEN")

  # Create Hugging Face Endpoint object (defines the model and task)
  llm = HuggingFaceEndpoint(
      repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
      task="text-generation",
      huggingfacehub_api_token=hf_token,
  )

  # Create ChatHuggingFace object using the endpoint
  chat_model = ChatHuggingFace(llm=llm)

  # Invoke the chat model
  result = chat_model.invoke([HumanMessage(content="What is the capital of India?")])

  # Print the content
  print(result.content)
  ```
             

## Model Parameters
- Two important parameters for language models are **Temperature** and **Max Completion Tokens**.    
- **Temperature** controls the randomness/creativity of the output (0 to ~2). Lower values (0-0.3) yield deterministic/predictable responses, while higher values (1.5+) result in more random/creative/diverse responses.     
- **Max Completion Tokens** limits the maximum number of tokens (roughly words) in the output. This is useful for controlling costs with paid models.     

| Temperature Range | Recommended Use Case                                   |
| :---------------- | :----------------------------------------------------- |
| **0 - 0.3**       | Factual answers, Math, Code generation                 |
| **0.5 - 0.7**     | General Q&A, Explanations                              |
| **0.9 - 1.2**     | Creative writing, Storytelling, Jokes                  |
| **1.5+**          | Highly random/creative outputs, Brainstorming          |
   

## Using Embedding Models
- Embedding models convert text into numerical vectors (embeddings) to capture contextual understanding.   
- They are primarily used for **semantic search** and building **RAG-based applications**.    
- You can use closed-source embedding models like OpenAI's.   
- Example using OpenAI Embeddings:
  ```python
  from langchain_openai import OpenAIEmbeddings
  from dotenv import load_dotenv
  import os

  load_dotenv()

  # Create Embedding Model object
  embedding_model = OpenAIEmbeddings(
      model="text-embedding-3-large", # Choose the model
      dimensions=300 # Specify vector dimensions (optional, affects cost/context)
  )

  # Embed a single query
  query_embedding = embedding_model.embed_query("Delhi is the capital of India")
  print(str(query_embedding))

  # Embed multiple documents
  documents = [
      "Delhi is the capital of India.",
      "Kolkata is the capital of West Bengal.",
      "Paris is the capital of France."
  ]
  doc_embeddings = embedding_model.embed_documents(documents)
  # doc_embeddings will be a 2D list
  print(str(doc_embeddings))
  ```
                  
- OpenAI embedding costs are very low (around $0.002 per 1 million tokens).    
- You can also use open-source embedding models, often by downloading them locally.   
- Hugging Face is the largest repository for open-source LLMs and embedding models.

## Document Similarity Application
- Embedding models are useful for finding the similarity between documents or a query and documents.   
- The process involves:
    1. Generating embeddings for a set of documents.    
    2. Generating an embedding for the user's query.    
    3. Calculating the similarity between the query embedding and each document embedding (e.g., using Cosine Similarity).     
    4. Identifying the document with the highest similarity score as the most relevant.    
- This is the fundamental concept behind RAG (Retrieval Augmented Generation) applications.    
- Example code uses `OpenAIEmbeddings` and `CosineSimilarity` from `sklearn.metrics.pairwise`:
  ```python
  from langchain_openai import OpenAIEmbeddings
  from dotenv import load_dotenv
  from sklearn.metrics.pairwise import cosine_similarity
  import numpy as np
  import os

  load_dotenv()

  # Create Embedding Model
  embedding_model = OpenAIEmbeddings(
      model="text-embedding-3-large",
      dimensions=300
  )

  # Sample Documents
  documents = [
      "Virat Kohli is an Indian cricketer known for his aggressive approach.",
      "Sachin Tendulkar is a legendary Indian cricketer, often called the 'Master Blaster'.",
      "Jasprit Bumrah is an Indian fast bowler known for his unique action.",
      "Rohit Sharma is the current captain of the Indian cricket team.",
      "MS Dhoni is a former Indian captain and wicket-keeper known for his finishing skills."
  ]

  # User Query
  query = "Tell me about Jasprit Bumrah"

  # Generate embeddings for documents
  doc_embeddings = embedding_model.embed_documents(documents)

  # Generate embedding for the query
  query_embedding = embedding_model.embed_query(query)

  # Reshape query embedding for cosine_similarity (needs 2D input)
  query_embedding_2d = np.array(query_embedding).reshape(1, -1)

  # Calculate similarity scores
  scores = cosine_similarity(query_embedding_2d, doc_embeddings)[0] # Get the 1D array of scores

  # Pair scores with original document index and sort
  indexed_scores = list(enumerate(scores))
  sorted_scores = sorted(indexed_scores, key=lambda x: x[1], reverse=True) # Sort by score descending

  # Get the index and score of the most similar document
  most_similar_index, highest_score = sorted_scores[0]

  # Print the query and the most similar document content and score
  print(f"Query: {query}")
  print(f"Most Similar Document: {documents[most_similar_index]}")
  print(f"Similarity Score: {highest_score}")
  ```

## Importance of Prompts

* Prompts are critically important for the output quality of LLMs, and various techniques are used for prompt construction.
* Due to the significance of prompts, a new job profile has emerged called **Prompt Engineering**.
* LangChain provides multiple techniques for designing prompts.

---

## Building a Simple Chatbot

* A simple console-based chatbot can be created that takes input from the user.
* The user's input is sent as a prompt to the LLM using the `model.invoke()` function.
* The result from the LLM (`result.content`) is then displayed on the console.

**Problem:**
This basic chatbot does **not remember previous messages (context)**.
Each new input is treated as an independent query by the LLM, which can lead to incorrect responses for context-based follow-up questions.

---

## Chat History and Message Types

* To solve the context problem, a **chat history** (a list of messages) should be maintained.
* Each user input and AI response is appended to the chat history, and the full history is passed to `model.invoke()`.
* The `model.invoke()` function can accept either a single message or a list of messages (chat history).
* **Problem:** A basic list of messages doesn‚Äôt indicate who said what ‚Äî user or AI.
* Without labeling messages, the LLM may get confused during long conversations, leading to misinterpretation.

To solve this, **LangChain provides three types of messages**:

* **System Message**: Sent at the start of a conversation to instruct the LLM (e.g., "You are a helpful assistant").
* **Human Message**: Represents messages sent by the user.
* **AI Message**: Represents responses from the AI.

Using these message types helps the LLM better understand the **flow and context** of a conversation.

---

## Static vs Dynamic Prompts

Prompts can be either **Static** or **Dynamic**.

### Static Prompts

* These are hardcoded by the developer directly into the code.
* **Problem:** Static prompts give users full freedom to write the prompt, which reduces control over the LLM output.
* Even a slight change in user input can significantly alter the LLM response.
* If a prompt is unclear or incorrect, the LLM might generate wrong or **hallucinated** responses.
* Static prompts are **not ideal** for consistent user experience in LLM applications.

### Dynamic Prompts

* These use **templates** with placeholders (fill-in-the-blanks).
* Values for placeholders are filled at runtime (typically from user input or other variables).
* Allows one template to generate multiple prompts for different inputs.
* Provides better control over LLM output and ensures consistency.

| Feature                 | Static Prompts                       | Dynamic Prompts                                      |
| ----------------------- | ------------------------------------ | ---------------------------------------------------- |
| **Definition**          | Hardcoded by the developer           | Template with placeholders                           |
| **User Control**        | High (user writes the whole prompt)  | Limited (user only provides values for placeholders) |
| **Consistency**         | Low (depends on user prompt quality) | High (template enforces structure)                   |
| **Error/Hallucination** | High risk                            | Low risk (inputs are guided via templates)           |
| **Use Case**            | Simple, specific tasks               | Real-world apps, tools                               |

## `PromptTemplate` in LangChain

* In LangChain, the **`PromptTemplate`** class is used to create dynamic prompts for **single-turn** queries.
* `PromptTemplate` takes a prompt string template and input variables (placeholder names).
* The `template.invoke()` method is used to fill the placeholders using a dictionary.

**Advantages over f-strings:**

* **Validation**: Ensures that all required variables are provided and no extras are passed, helping avoid runtime errors.
* **Reusability**: Prompt templates can be saved as JSON files (`template.save()`) and loaded later (`load_prompt()`).
* **Integration**: Works seamlessly with other LangChain components like **Chains**.

---

## `ChatPromptTemplate` for Multi-Turn

* When working with **multi-turn conversations** (a list of messages), the `ChatPromptTemplate` class is used to build dynamic prompts.
* It lets you define system, human, and AI messages within a message list.
* The template accepts a list of message templates.
* Each message template is passed as a tuple: `(role, content_string)`.
* The `role` (like 'system', 'human', or 'ai') defines the type of message.
* The `content_string` can contain placeholders filled during runtime using `chat_template.invoke()`.
* The purpose of `ChatPromptTemplate` is similar to `PromptTemplate`, but designed specifically for multi-turn chat scenarios.

---

## Understanding Message Placeholders in LangChain

* A **Message Placeholder** in LangChain is a special type of placeholder used within a **Chat Prompt Template**.
* Its main role is to **dynamically insert a chat history (list of messages)** into the template at runtime.
* While you can manually pass a list of messages to an LLM, using a Message Placeholder simplifies this process within templated chat interactions.

---

## Purpose and Problem Solved

* Message Placeholders help maintain **context** in multi-turn chats (like in a chatbot).
* For example, in customer support, a user might return later with a follow-up query about their refund.
* The chatbot needs access to the previous conversation to understand the new question correctly.
* That conversation history can be **stored** (e.g., in a database) and **reloaded** later.
* Message Placeholders make it easy to dynamically inject that past history into the current prompt, preserving context.

---

## How Message Placeholders Work

* You define a placeholder within the `ChatPromptTemplate` specifically for chat history.
* This acts as a "slot" where the list of previous messages is inserted.
* When invoking the template, you pass the actual chat history as a list of messages to this placeholder variable.
* The final prompt generated includes:

  1. System message
  2. Chat history
  3. Current user input
* This helps the LLM understand the full context and respond accordingly.

## Understanding Structured Output in LangChain
- **Structured output** refers to having Language Models (LLMs) return responses in a well-defined data format, such as JSON, making the output easier to parse and work with programmatically.    
- Traditionally, LLMs produce **unstructured output**, which is typically text without a defined format, making it difficult for other systems to process.    
- By enabling structured output, LLMs can integrate with other systems like databases and APIs, which was previously challenging due to the textual nature of LLM responses.    

  

- This capability is crucial for building advanced applications like **Agents**, which require tools to receive structured input to perform tasks.    

## Key Use Cases for Structured Output
- **Data Extraction**: Extracting specific information from unstructured text (e.g., resume details like name, marks, last company) and storing it in a database.     
- **API Building**: Creating APIs that process unstructured data (e.g., phone reviews) and return structured insights like key topics, pros, cons, and overall sentiment.      
- **Building Agents**: Agents, which are "chatbots on steroids" capable of performing actions, rely on structured output to feed specific, formatted data to their tools (e.g., sending numbers to a calculator tool).      

  

## Methods for Generating Structured Output in LangChain
- LangChain provides the `with_structured_output` function to simplify generating structured output from LLMs that support it, such as OpenAI&apos;s GPT models.    
- For models that do not inherently support structured output, **Output Parsers** in LangChain can convert unstructured output into a desired structure.    
- Three primary methods exist for specifying the data format (schema) in LangChain: **Typed Dictionary**, **Pydantic**, and **JSON Schema**.    

  

### Typed Dictionary
- A **Typed Dictionary** in Python defines a dictionary&apos;s structure by specifying the expected keys and their value data types, helping to ensure a specific structure is followed.    
- It improves collaboration among programmers by communicating expected data types, but it does not perform runtime validation; code will still run even if types are mismatched.      
- You define a `TypedDict` by creating a class that inherits from `TypedDict` and declaring keys with their type hints (e.g., `Name: str`, `Age: int`).    
- **Annotations** can be added to provide more detailed descriptions to the LLM about what kind of data to generate for a specific field.     
- Fields can be marked as **optional** using `Optional` from the `typing` module, allowing the LLM to omit them if the information is not present.   
- **Literal** types can restrict a field&apos;s value to a predefined set of strings (e.g., &quot;Positive&quot;, &quot;Negative&quot;, &quot;Neutral&quot;).   
- When using `with_structured_output` with Typed Dictionary, a system prompt is generated behind the scenes to guide the LLM to produce output in the desired JSON format.      

### Pydantic
- **Pydantic** is a Python library for data validation and parsing, allowing developers to enforce criteria on incoming data.   
- Unlike Typed Dictionary, Pydantic **validates data at runtime** and raises errors if the data does not conform to the defined schema, making it suitable for APIs and data processing.     
- Schemas are defined by creating classes that inherit from Pydantic&apos;s `BaseModel`, specifying attributes with their expected types.    
- Pydantic supports **default values** for fields, which are used if no value is provided.    
- It handles **optional fields** similarly to Typed Dictionary, requiring `Optional` from `typing` and a `None` default if the value is not set.     
- Pydantic performs **type coercion** (implicit data conversion) where it automatically attempts to convert input data to the expected type if possible (e.g., string &quot;32&quot; to integer 32).     
- It includes **built-in validations** for common data types like email addresses (`EmailString`).     
- The **`Field` function** provides advanced options for fields, including:
    -   Setting default values.   
    -   Applying **constraints** (e.g., numerical ranges like 0 to 10 for CGPA).     
    -   Adding **descriptions** for LLM guidance, similar to Typed Dictionary annotations.    
    -   Attaching **regular expressions** for pattern validation.   
- Pydantic objects can be easily converted to Python dictionaries or JSON strings for integration with other systems.     

### JSON Schema
- **JSON Schema** is used when a project involves **multiple programming languages** (e.g., Python backend, JavaScript frontend) and the schema needs to be universally understood.     
- It defines the structure, data types, and required properties of a JSON object.      
- Key components of a JSON Schema include `title`, `description`, `type` (e.g., &quot;object&quot;, &quot;array&quot;), `properties` (for attributes), and `required` fields.      
- For lists, `type` is &quot;array&quot; and `items` specifies the type of elements within the array.   
- For literal options, `enum` is used instead of `Literal`.   
- Optional fields are indicated by allowing `null` as one of the possible types for a property (e.g., `type: [&quot;string&quot;, &quot;null&quot;]`).    
- When used with `with_structured_output`, the function returns a Python dictionary, similar to Typed Dictionary.   

## Choosing the Right Method
- **Typed Dictionary** is suitable when working exclusively in Python and only needing **type hints** without runtime validation or sharing the schema with other languages.   
- **Pydantic** is the recommended &quot;go-to&quot; format for most Python-based projects, especially when **data validation**, default values, or automatic type conversion are required.     
- **JSON Schema** is ideal for **cross-language compatibility** and when the schema needs to be shared and understood across different programming environments.   

| Feature / Method       | Typed Dictionary | Pydantic        | JSON Schema     |
| :--------------------- | :--------------- | :-------------- | :-------------- |
| **Python-only**        | Yes              | Yes             | No              |
| **Type Hints**         | Yes              | Yes             | No              |
| **Runtime Validation** | No               | Yes             | No (inherently) |
| **Default Values**     | No (direct)      | Yes             | No              |
| **Auto Type Conversion** | No               | Yes             | No              |
| **Cross-Language Comp.** | No               | No              | Yes             |
     

## `with_structured_output` Function Parameters
- The `with_structured_output` function includes a `method` parameter that defines how the structured output is generated.   
- **`json_mode`**: Used when the desired structured output is in JSON format, typically for models like Claude or Gemini.    
- **`function_calling`**: This is the default method for OpenAI models and is used when the structured output is intended to invoke a specific function or tool, common in agentic workflows.     
- Some models, like TinyLlama, do not support either `json_mode` or `function_calling` for structured output, requiring the use of **Output Parsers** instead.

## Introduction to Output Parsers

* In **LangChain**, **Output Parsers** help convert raw (textual) LLM responses into structured formats like JSON, CSV, or Pydantic models.
* They ensure **consistency, validation, and ease of use** in applications.
* These are classes written in LangChain that let you work with any type of LLM‚Äîwhether or not they natively support structured output.
* This video focuses on the **four most important output parsers**: String, JSON, Structured, and Pydantic.

## Types of LLMs and Structured Output

* LLMs can be broadly classified into two categories based on their ability to produce structured output:

  * **Models that can produce structured output**: These are fine-tuned to give structured responses.
  * **Models that cannot produce structured output**: These are typically open-source models not fine-tuned for structured responses.
* LangChain provides the `with_structured_output` function to work with models that naturally support structured output.
* For models that don‚Äôt support structured output by default, **output parsers** become crucial.

## String Output Parser

* The **String Output Parser** is the simplest, converting LLM responses into a plain string format.
* LLM responses often contain additional **metadata** (like token usage) beyond the actual textual content.
* Using this parser removes the need to manually extract the `result.content` attribute.
* It's especially useful for **chaining multiple LLM calls**, where one call‚Äôs output becomes the next call‚Äôs input‚Äîensuring clean text transfer.
* Commonly used in **LangChain Chains** to create operation pipelines.

## JSON Output Parser

* The **JSON Output Parser** forces the LLM to return its output in **JSON format**.
* This is the fastest way to get JSON output from an LLM.
* When using this parser, you send additional instructions to the LLM via `parser.get_format_instructions()` to specify the desired JSON structure.
* The main **limitation** is that it does **not enforce a schema**; the structure of the JSON object is up to the LLM.
* You can‚Äôt strictly specify key-value structures (e.g., `Fact One: "..."`, `Fact Two: "..."`) and there‚Äôs no guarantee they will be returned.

## Structured Output Parser

* The **Structured Output Parser** extracts **structured JSON data** from an LLM response based on a predefined schema.
* Its biggest advantage is that it **enforces a schema**, guiding the LLM to generate output in a specified structure.
* The schema is created using the `ResponseSchema` class, usually as a list of schema objects.
* A primary **drawback** is its **inability to validate data**.
* For example, if you request an integer age but receive "35 years" (a string), this parser cannot block or correct it.

## Pydantic Output Parser

* The **Pydantic Output Parser** is a structured output parser that applies schema enforcement and **data validation** on the LLM response using **Pydantic models**.
* It offers **strict schema enforcement** (data types, constraints), **type safety**, **easy validation**, and **seamless integration**.
* You define a schema using a Pydantic `BaseModel` class, specifying attributes with data types (e.g., `name: str`, `age: int`) and **constraints** (e.g., `age > 18`).
* This parser automatically generates the required formatting instructions for the LLM based on the Pydantic model.
* It produces **structured JSON** and validates the data against the defined Pydantic schema.

Example Pydantic model definition:

```python
class Person(BaseModel):
    name: str = Field(description="Name of the person")
    age: int = Field(description="Age of the person", gt=18)
    city: str = Field(description="City name of the city the person belongs to")
```

## Output Parser Summary

| Parser Type           | Primary Function                                     | Schema Enforcement | Data Validation | Main Use Case                                                    |
| --------------------- | ---------------------------------------------------- | ------------------ | --------------- | ---------------------------------------------------------------- |
| **String Output**     | Converts LLM response into plain string              | No                 | No              | Chaining LLM calls, simple text extraction                       |
| **JSON Output**       | Forces LLM to return output in JSON format           | No                 | No              | Quick JSON output when structure isn't critical                  |
| **Structured Output** | Extracts structured JSON based on predefined schema  | Yes                | No              | When specific JSON structure is needed without strict data types |

## 1. Models and Prompts

- **Models:** In LangChain, models are components that allow you to select and interact with various types of AI or LLMs (Large Language Models).
- **Prompts:** Prompts are used to send different types of input to the LLM so that it can produce the desired output.
- **Structured Output & Parsers:** LangChain helps in obtaining structured and consistent output from LLMs. For this, output parsers are used to convert output data into a particular format or structure.

## 2. Chains: Need and Working

- LLM-based applications usually work through a series of small steps, such as:
  - Taking input from the user
  - Sending it to the LLM as a prompt
  - Processing the model's response
- Performing all these steps manually again and again can be complex and time-consuming, especially for complicated applications.
- The concept of **Chains** addresses this; it automates these steps by connecting them in a pipeline fashion.
- In the pipeline, the output of one stage becomes the input for the next, and the entire process operates without manual intervention.
- A major benefit of Chains is that the user provides input just once, and after complete processing, the final output is received directly.

## 3. Benefits and Types of Chains

- Chains support not only sequential (linear) flow, but also parallel and conditional flows.
- In LangChain, the main focus is on three types of chains:
  - **Sequential Chain**
  - **Parallel Chain**
  - **Conditional Chain**

## 4. Building a Simple Sequential Chain

- A simple chain usually consists of three steps: prompting, model calling, and output parsing.
- Using **LangChain Expression Language (LCEL)** and the pipe (`|`) operator, chains are defined declaratively, for example:
  ```
  prompt | model | parser
  ```
- Once input is given, the chain automatically moves through all pipeline steps.
- This makes the code simpler and eliminates the need for repeated `invoke` calls for each step.
- You can visualize the chain structure using the `chain.get_graph().print_ascii()` function.

## 5. Building a Complex Sequential Chain

- Complex chains may have multiple LLM calls.
- Example: First, generating a report on a topic using an LLM, then extracting the main points from that report with another LLM call.
- Such a chain could look like:
  ```
  prompt1 | model | parser | prompt2 | model | parser
  ```
- After each step, the output becomes the next input, making it easy to automate complex logic.

## 6. Building a Parallel Chain

- If you want to perform multiple tasks (like generating both notes and a quiz) from a single input, parallel chains are used.
- Using **RunnableParallel**, you can define different chains with unique names in a dictionary-like way, which run simultaneously.
- Each chain works independently on the same input, and their outputs can be merged later.
- RunnableParallel is a part of LangChain‚Äôs **Runnable** abstraction.

## 7. Building a Conditional Chain

- When you want to run different tasks based on a condition or output, conditional chains are used.
- With **RunnableBranch**, you can build chains that work like `if-else` logic.
- It consists of a list of tuples, each with a condition and a chain to execute if the condition is met; you can also define a default case.
- To ensure outputs are in the expected format, **PydanticOutputParser** is important.
- With **RunnableLambda**, lambda functions can be converted into Runnables‚Äîfor adding custom or default logic.

## 8. Importance of Runnable

- A key concept in LangChain is the **Runnable** abstraction.
- All chains‚Äîwhether Parallel, Branch, or Lambda‚Äîare implemented around this abstraction.
- Runnable is the foundation of LangChain‚Äôs pipelining and composability features.
- Further study of the Runnable abstraction will be covered in upcoming videos, providing in-depth understanding of LCEL examples and internal processing.

### The Evolution of LangChain: From Components to Chains
- Around November 2022, with the release of ChatGPT and OpenAI APIs, the LangChain team foresaw a high demand for **LLM-based applications**.     
- **Initial Problem (Inconsistent LLM APIs):** Different LLM providers (e.g., Anthropic, Google, Mistral) had varying API behaviors. LangChain addressed this by providing a unified framework to interact with any LLM API with minimal code changes.     
- **Second Problem (Beyond LLM Interaction):** Building LLM applications involves many tasks beyond just interacting with an LLM, such as document loading, splitting, embedding generation, vector database storage, retrieval, and output parsing.     
- LangChain evolved by creating dedicated components for each of these tasks, making it a powerful library for AI engineers to build complex LLM applications.

### The Rise and Fall of Chains
- LangChain observed common patterns in how developers connected components (e.g., creating a prompt and sending it to an LLM).    
- This led to the concept of **Chains**, which are built-in functions designed to automate and connect two or more components into a pipeline, simplifying development.      
- Examples of early Chains include:
    - **LLMChain:** Connects an LLM and a Prompt Template, automating prompt generation and LLM prediction.      
    - **RetrieverQA Chain:** Automates the Retrieval-Augmented Generation (RAG) process by connecting a retriever and an LLM.      
- LangChain developed numerous Chains for various reusable tasks, such as `SimpleSequentialChain`, `SequentialChain`, SQL database chains, API chains, and Math chains.      

---

- **Unintended Problem:** The proliferation of too many Chains resulted in a large, difficult-to-maintain codebase and a steep learning curve for new AI engineers.       
- This issue arose because the initial LangChain components (LLM, Prompt, Parsers, Retrievers) were **not standardized**; they had different interaction methods (e.g., `predict` for LLM, `format` for Prompt Template).      
- The lack of standardization forced LangChain to write custom Chain functions for every new use case to bridge these incompatibilities.       

### The Solution: Runnables
- LangChain recognized the need to rebuild components to be **standardized** and **seamlessly connectable** using **Runnables**.

- **Lego Blocks Analogy:** Runnables are like Lego blocks, each a unit of work, following a common interface, connectable, and forming larger structures that are themselves Lego blocks.        

### Implementing Runnables (Code Walkthrough)
- To standardize components, an **abstract `Runnable` class** is created, defining an abstract `invoke` method.      
- All component classes (e.g., `NakliLLM`, `NakliPromptTemplate`, `NakliStrOutputParser`) inherit from `Runnable` and are forced to implement the `invoke` method, ensuring a common interface.         
- Old methods like `predict` (for LLM) and `format` (for Prompt Template) are marked for deprecation, encouraging the use of `invoke`.     
- A **`RunnableConnector` class** is used to compose Runnables into chains. It takes a list of Runnables and its `invoke` method iterates through them, passing the output of one Runnable as the input to the next.         
- This design allows for the creation of arbitrarily long and complex chains without writing custom functions for each connection.     
- **Chaining Chains:** Multiple chains, which are themselves Runnables, can be connected to form even larger, more complex workflows.

### Types of Runnables
- Runnables in LangChain can be divided into two main categories: **Task-Specific Runnables** and **Runnable Primitives**.

- **Task-Specific Runnables**: These are the core LangChain components that have been converted into runnables for use in pipelines.
    - Each has a specific purpose, such as prompt templates for designing prompts and the LLM component for interacting with the LLM.
    - Examples include `ChatOpenAI`, `PromptTemplate`, and `Retriever`.
- **Runnable Primitives**: These are runnables that help connect other task-specific runnables, enabling the creation of complex AI workflows.
    - They are the fundamental building blocks for structuring execution logic in AI workflows, defining how different runnables interact sequentially, in parallel, or conditionally.

### LangChain Expression Language (LCEL)
- **LCEL** is a **declarative way** to define chains in LangChain.
- The creators of LangChain noticed that creating sequential chains using `RunnableSequence` was a very specific use case.
- Therefore, a syntax like `RunnableSequence(R1, R2, R3)` has been replaced with a simpler syntax like `R1 | R2 | R3` (the pipe operator).
- Currently, LCEL is primarily used for `RunnableSequence`.

## Introduction to RAG-based Applications
- RAG (Retrieval Augmented Generation) applications combine information retrieval with language generation to provide accurate and grounded responses.   
- They address limitations of LLMs, such as not having up-to-date information or access to personal/company-specific data.      
- Benefits of RAG include providing up-to-date information, enhanced privacy by not requiring confidential data uploads, and the ability to process large documents by dividing them into chunks.

## Document Loaders: Purpose and Structure
- **Document Loaders** are components in LangChain used to load data from various sources into a standardized format, typically as **Document objects**.   
- This standardization ensures that data can be easily used with other LangChain components for tasks like chunking, embedding, retrieval, and generation.    
- All document loaders are found within the `langchain_community.document_loaders` package.    
- Each **Document object** comprises two main parts: **page_content**, which holds the actual data, and **metadata**, containing information such as the source, creation date, or author.    
- Regardless of the specific loader used, every LangChain document loader returns a Python **list of Document objects**.     

## Key Document Loaders in LangChain

### TextLoader
- The **TextLoader** is the simplest document loader, designed to load **text files** and convert them into Document objects.   
- It is initialized with the file path and can optionally specify the encoding (e.g., `utf8`) for special characters.    
- The `loader.load()` method reads the text file and returns a list containing a single Document object.     
- You can access the text content via `doc.page_content` and associated information through `doc.metadata`.   

### PyPDFLoader
- The **PyPDFLoader** reads **PDF files** and converts them into Document objects.   
- It operates on a page-by-page basis, meaning a 25-page PDF will yield a list of 25 Document objects, each representing one page.   
- Each Document object includes its page content and metadata, such as the page number and source.   
- This loader utilizes the `PyPDF` library internally and is best suited for PDFs primarily containing textual data.    
- It may not perform optimally with **scanned PDFs** or complex, image-heavy PDF files.    

### DirectoryLoader
- The **DirectoryLoader** is used to load multiple files from a specified directory.   
- It requires a directory path, a `glob` pattern (e.g., `*.pdf`) to filter files, and a `loader_cls` (the specific document loader to apply, such as `PyPDFLoader`).     
- For instance, loading three PDF books with a combined total of 1186 pages would result in 1186 Document objects.   

### Lazy Loading vs. Eager Loading
- LangChain document loaders provide two primary methods for loading documents: `load()` (eager loading) and `lazy_load()` (lazy loading).   

| Feature         | `load()` (Eager Loading)                               | `lazy_load()` (Lazy Loading)                                |
|:-------------- |:----------------------------------------------------- |:---------------------------------------------------------- |
| **Behavior**    | Loads all documents into memory at once.               | Loads documents on demand, one at a time.                   |
| **Return Type** | Returns a **list of Document objects**.                | Returns a **generator of Document objects**.                |
| **Memory Usage**| High; all documents reside in RAM simultaneously.      | Low; only one document is in memory at a time.              |
| **Use Case**    | Suitable for a small number of documents or small files where all data is needed upfront.    | Ideal for a large number of documents or very large files, or when stream processing is required.     |
| **Performance** | Initial delay as all documents are processed before any output.    | Immediate processing and output as documents are yielded one by one.     |
              

### WebBaseLoader
- The **WebBaseLoader** extracts text content from **web pages** and loads it into LangChain.    
- It internally uses the `requests` library for HTTP requests and `BeautifulSoup` for parsing HTML to extract text.   
- This loader is effective for static websites, blogs, news articles, or public websites.    
- Its limitation is that it may not work well with JavaScript-heavy web pages; for such cases, `SeleniumURLLoader` is a better alternative.    
- It returns a single Document object for a single URL but can also accept a list of URLs to load multiple web pages.    

### CSVLoader
- The **CSVLoader** is used to load **CSV files** into LangChain.   
- For every row in the CSV file, it creates a separate Document object.   
- Each Document object's `page_content` contains a string representation of the row's column names and values, while `metadata` includes the source and row number.    
- It supports lazy loading for large CSV files, enabling row-by-row processing.   

## Other Document Loaders and Custom Loaders
- LangChain offers hundreds of document loaders, categorized by data source (e.g., web pages, cloud services, social platforms, common file types like JSON).       
- It is not necessary to learn all of them; instead, it is recommended to learn on a project-by-project basis by referring to the documentation.    
- LangChain allows users to create **custom document loaders** by inheriting from the `BaseLoader` class and implementing their own `load()` and `lazy_load()` functions.    
- Many existing LangChain document loaders were developed by the community using this custom loader feature.

## Text Splitters: An Essential Component for RAG Applications
- Text splitting is the process of breaking down large texts, such as articles, PDFs, HTML pages, or books, into smaller, manageable pieces that can be effectively handled by **Large Language Models (LLMs)**.
- This process is crucial for building Retrieval-Augmented Generation (RAG) based applications, which involve important components like document loaders and text splitters.

### Why Text Splitting is Important for LLM-Powered Applications
- Text splitting is vital for LLM-powered applications due to three main reasons:
      

- **Overcoming Model Limitations**: LLMs have a **context length limit**, meaning they can only process a certain amount of text input at once (e.g., 50,000 tokens). Text splitting allows processing documents that would otherwise exceed these limits, enabling tasks like summarizing large PDFs.
- **Improving Downstream Tasks**: Text splitting leads to better results for various LLM-powered tasks:
    - **Embedding Quality**: When embedding large texts into vectors, the semantic meaning might not be well-captured. Splitting text into smaller chunks and embedding each chunk separately results in better semantic meaning capture.
    - **Semantic Search**: Performing semantic search after chunking or text splitting yields more precise and improved search quality compared to searching on a large, un-split text.
    - **Summarization**: LLMs can sometimes drift or hallucinate when summarizing very large documents. Text splitting has been empirically proven to provide better summarization results in such situations.
- **Optimizing Computational Resources**: Working with smaller text chunks is more memory-efficient and allows for better parallelization of processing tasks, reducing computational requirements.

### Types of Text Splitters
- LangChain supports four main types of text splitters: length-based, text structure-based, document structure-based, and semantic meaning-based.

---
#### Length-Based Text Splitting (Character Text Splitter)
- This is the simplest and fastest method, where text is split based on a predefined chunk size, typically measured in characters or tokens.
- The process involves traversing the text from the beginning, creating a chunk once the specified character count is reached, and then resuming from that point for the next chunk.
- **Advantages**: It is conceptually simple, easy to implement, and very fast.
- **Disadvantages**: It does not consider linguistic structure, grammar, or semantic meaning, often cutting words, paragraphs, or sentences in the middle, which can lead to loss of context and poor embedding quality.
- **Implementation in LangChain**: The `CharacterTextSplitter` class is used, where `chunk_size` and `chunk_overlap` parameters are specified.
    - The `split_text` function is used for plain text, while `split_documents` is used for document objects loaded by document loaders.

##### Chunk Overlap
- **Chunk overlap** defines the number of characters that overlap between two consecutive chunks.
- **Purpose**: It helps retain context that might be lost when text is abruptly cut by the splitter. By starting the next chunk slightly before the previous one ended, some information is shared, preventing context loss.
- **Trade-off**: While more overlap helps retain context, it also increases the number of chunks, leading to more computational overhead.
- **Recommended Overlap**: For RAG-based applications, a chunk overlap of 10% to 20% of the chunk size is generally considered good (e.g., 10-20 characters for a 100-character chunk).

---
#### Text Structure-Based Text Splitting (Recursive Character Text Splitter)
- This technique considers the inherent hierarchical structure of text (paragraphs, sentences, words, characters).
- The **Recursive Character Text Splitter** is one of the most widely used text splitting techniques in LangChain.
- **How it works**: It attempts to split text using a predefined list of separators in a recursive manner, prioritizing larger structural units first:
    1.  It first tries to split by **paragraphs** (`\n\n`).
    2.  If chunks are still too large, it tries to split by **lines/sentences** (`\n`).
    3.  If still too large, it tries to split by **words** (` `).
    4.  Finally, if necessary, it splits by **characters** (nothing).
- This algorithm continuously tries to avoid abrupt mid-text splits, ensuring that if a larger chunk size is allowed, it will break at logical points like paragraph or sentence boundaries.
- **Implementation in LangChain**: The `RecursiveCharacterTextSplitter` class is used, with `chunk_size` and `chunk_overlap` parameters.
- This splitter is preferred over the character splitter because it intelligently breaks down text based on context, making it more suitable for LLM applications.

---
#### Document-Based Text Splitting
- This is an extension of the recursive character text splitter, designed for documents that are not plain text but follow specific formatting rules, such as code or Markdown files.
- Instead of standard paragraph/sentence separators, this technique uses **different sets of separators** tailored to the document type (e.g., `class` and `def` keywords for Python code, or headings for Markdown).
- The underlying algorithm remains the same as the recursive character text splitter, but the separators are customized to respect the document's unique structure.
- **Implementation in LangChain**: The `RecursiveCharacterTextSplitter.from_language` method is used, where you specify the `Language` (e.g., `Language.PYTHON`, `Language.MARKDOWN`, `Language.HTML`) along with `chunk_size` and `chunk_overlap`.

---
#### Semantic Meaning-Based Text Splitting
- This advanced technique makes splitting decisions based on the **semantic meaning** of the text, rather than just length or structural cues.
- **How it works**:
    1.  The text is first broken down into smaller units, typically sentences.
    2.  An **embedding model** (e.g., OpenAI's embedding model) is used to generate embedding vectors for each sentence.
    3.  The **similarity** (e.g., cosine similarity) between consecutive sentence embeddings is calculated.
    4.  A **sliding window approach** is used to compare similarities. When a sudden, significant drop in similarity is detected between two consecutive sentences, it indicates a topic change, and a split is performed at that point.
    5.  A **threshold type** (e.g., standard deviation, percentile, interquartile range, gradient) and amount are used to determine what constitutes a "significant drop" in similarity.
- **Current Status**: This concept is relatively new and the text splitters based on semantic meaning are currently **experimental** in LangChain, and their performance may not always be accurate.
- **Future Potential**: As embedding models become more powerful, this technique is expected to become more prevalent in the future.
- **Implementation in LangChain**: The `SemanticChunker` class from `langchain_experimental` is used, requiring an embedding model and a `threshold_type`.

## Introduction to Vector Stores

## Semantic Search with Embeddings
- A better approach is to compare the **plots or stories** of movies to determine similarity.    
- This requires storing movie plots in the database and a system to compare them.    
- **Embeddings** are a deep learning technique that represents the semantic meaning of text (like movie plots) as numerical vectors.     
- Once text is converted to vectors, the similarity between two pieces of text can be calculated by finding the **angular distance** between their corresponding vectors in a multi-dimensional coordinate system; a smaller angular distance indicates higher similarity.     

## Challenges in Building Semantic Search Systems
- Building such a system presents three main challenges:
    1.  **Generating Embeddings:** Creating embedding vectors for potentially millions of documents.   
    2.  **Storage:** Storing these embedding vectors efficiently, as traditional relational databases (like MySQL) are not suitable for vector operations.    
    3.  **Efficient Semantic Search:** Performing fast similarity searches across a large number of vectors (e.g., finding the most similar movie out of 10 lakh movies without comparing each one linearly, which is computationally expensive and slow).      
- **Vector stores** are designed to solve these three challenges.   

## Key Features of Vector Stores
- Vector stores are systems built to store and retrieve data represented as numerical vectors.    
- They offer four core features:
    -   **Storage:** They provide options to store vectors and their associated **metadata** (e.g., movie ID, name) either **in-memory** (for quick lookups, non-persistent) or **on-disk** (for durability and large-scale use, persistent).         
    -   **Similarity Search:** They enable calculating a similarity score between a query vector and all stored vectors to retrieve the most similar ones.    
    -   **Indexing:** They use data structures or methods to enable **fast similarity searches** on high-dimensional vectors, avoiding slow linear searches.       
        -   One indexing technique involves **clustering** vectors (e.g., 10 lakh vectors into 10 clusters), calculating centroids for each cluster, and then comparing a query vector only with the centroids. Once the closest centroid is found, the search is narrowed down to only the vectors within that specific cluster, significantly reducing comparisons (e.g., from 10 lakh to 1 lakh + 10).        
        -   Another technique is **Approximate Nearest Neighbor (ANN) lookup)**.   
    -   **CRUD Operations:** They support standard database operations: **C**reate (add new vectors), **R**etrieve (get existing vectors), **U**pdate (modify existing vectors), and **D**elete (remove outdated vectors).    
- Common use cases for vector stores include recommender systems, semantic search applications, RAG, and multimedia search.    

## Vector Store vs. Vector Database
- The terms **vector store** and **vector database** are often used interchangeably, but there's a distinction.    
- A **vector store** is typically a lightweight library or service focused primarily on storing vectors and performing similarity searches. It may lack traditional database features.    
    -   Example: **Faiss** (Facebook AI Similarity Search).   
- A **vector database** is a full-fledged database system designed to store and query vectors, offering additional database-like features.   
- The relationship can be summarized as: **Vector Store + Database-like Features = Vector Database**.    
- Database-like features include:
    -   **Distributed architecture** for scalability.    
    -   **Durability and persistence** (e.g., backup and restore).    
    -   **ACID transactions** (Atomicity, Consistency, Isolation, Durability) or near-ACID guarantees.    
    -   **Concurrency control** for multiple users.   
    -   **Authentication and authorization** for security.    
- Vector databases are ideal for production environments with significant scaling needs and large datasets.   
    -   Examples: **Milvus**, **Qdrant**, **Weaviate**, **Pinecone**.   
- Every vector database is a vector store, but not every vector store is a vector database.

    -   **Metadata filtering:** Filtering search results based on metadata attributes (e.g., `team` in cricket player example).     
    -   **Updating documents:** Using `update_document` with a document ID and the updated document object.      
    -   **Deleting documents:** Using the `delete` function with document IDs.    
- The interface for ChromaDB in LangChain is similar to other vector stores, making it easy to switch between them.

***RAG***

## Introduction to Retrieval Augmented Generation (RAG)
- RAG is a widely used and useful application in generative AI, combining language models with external knowledge retrieval to improve responses. It addresses limitations of large language models (LLMs) by augmenting them with relevant external information at query time.   
- The learning path involves understanding key components like document loaders, text splitters, vector stores, and retrievers, which are foundational to building RAG systems.    

## Limitations of Large Language Models (LLMs)
- LLMs are transformer-based neural networks with billions of parameters storing "parametric knowledge" learned during pre-training on massive datasets, mostly from the internet.    
- Users query LLMs by sending prompts, which the model answers based on its stored parametric knowledge.   
- However, LLMs face three main problems:
  1. They cannot answer questions about private data not seen during pre-training.    
  2. They have a knowledge cut-off date, so they cannot provide answers about recent events or data after that date.    
  3. They sometimes hallucinate, generating confident but factually incorrect information due to their probabilistic nature.     

## Fine-Tuning as a Solution and Its Challenges
- Fine-tuning involves retraining a pre-trained LLM on a smaller, domain-specific labeled dataset to incorporate specialized knowledge, improving performance on specific tasks.    
- Example analogy: An engineering graduate (pre-trained model) undergoes company-specific training (fine-tuning) to work effectively in a particular job.   
- Fine-tuning methods include supervised fine-tuning with labeled prompt-output pairs and unsupervised continued pre-training on domain data without labels.     
- Challenges of fine-tuning:
  - Computationally expensive and requires strong technical expertise and resources.   
  - Frequent updates to domain data require repeated fine-tuning, increasing cost and complexity.    
  - Managing removal or updates of outdated knowledge in the model is difficult.  

## In-Context Learning as an Alternative
- In-context learning is an emergent property of large LLMs (e.g., GPT-3 with 175 billion parameters) where the model learns to solve tasks by seeing examples within the prompt without updating weights.     
- This technique, also called few-shot prompting, allows the model to infer task instructions from examples and apply them to new inputs.     
- It mimics human learning from a few examples and reduces the need for costly fine-tuning datasets.     
- However, in-context learning may not work equally well for all tasks and often requires further alignment techniques like reinforcement learning with human feedback (RLHF) to improve performance.   

## Retrieval Augmented Generation (RAG) Concept
- RAG enhances LLMs by injecting relevant external context into the prompt at query time, combining parametric knowledge with up-to-date, domain-specific information.   
- Instead of sending only the user query, RAG sends both the query and retrieved context (e.g., relevant transcript segments) to the LLM, which then generates grounded and accurate responses.    
- This approach reduces hallucination by instructing the model to answer only from the provided context or say "I don't know" if insufficient information is available.   

## RAG Architecture and Workflow
- RAG combines two core concepts: information retrieval (a traditional computer science domain) and text generation (enabled by LLMs). It can be divided into four main steps:
  1. **Indexing:** Preparing an external knowledge base by loading documents, splitting large texts into meaningful chunks, generating embeddings (dense vector representations), and storing them in a vector database for efficient search.      
  2. **Retrieval:** At query time, the system converts the user query into an embedding vector, searches the vector store for the most relevant chunks, ranks them, and fetches the top results as context.     
  3. **Argumentation:** Combining the user query and retrieved context into a single prompt for the LLM, explicitly instructing it to answer only from the provided context.   
  4. **Generation:** The LLM reads the prompt, uses both its parametric knowledge and the injected context to generate a final response.    

  

## How RAG Solves LLM Limitations
- **Private Data Access:** By building the external knowledge base from private or proprietary data, RAG enables answering queries about private information without retraining the LLM.   
- **Recent Data Access:** New or updated documents can be added to the vector store without retraining the model, allowing RAG to provide answers based on the latest information.    
- **Reducing Hallucination:** Providing exact context and instructing the model to rely solely on it reduces the chance of hallucinated answers.    

## Advantages of RAG over Fine-Tuning
| Aspect                  | Fine-Tuning                          | RAG                                   |
|-------------------------|------------------------------------|-------------------------------------|
| Model Training          | Requires retraining the entire model| No retraining needed; uses existing LLM |
| Data Requirements       | Needs large labeled datasets       | Uses unlabeled documents as context |
| Cost and Complexity     | Computationally expensive and complex | Cheaper and simpler to implement    |
| Updating Knowledge      | Requires repeated fine-tuning      | Just update vector store with new data |
| Hallucination Control   | Limited, depends on training data  | Explicit context reduces hallucination |

  1. Writing a Python function that implements the task logic.

## Capabilities and Limitations of LLMs
- LLMs have two main powers: **reasoning** (understanding and breaking down questions) and **output generation** (producing answers based on parametric knowledge). However, LLMs cannot perform external tasks like modifying databases, posting on social media, or hitting APIs themselves. They can only generate text outputs, not execute actions. This limitation means LLMs lack "hands and feet" to perform tasks autonomously, which is a major challenge for making them powerful entities in software environments.

## Tools as a Solution
- The solution to LLMs' inability to perform tasks is to create **tools**‚Äîspecial functions that can interact with LLMs and perform specific tasks. Tools can be built-in or custom and act as intermediaries that LLMs can call upon to execute actions. Tools are essentially Python functions with capabilities to interact with LLMs when needed.

## Tool Binding: Connecting Tools with LLMs
- **Tool binding** is the process of registering tools with an LLM so the LLM knows which tools are available, what each tool does, and the input format required. This enables the LLM to call the appropriate tool with correctly formatted inputs during conversations. Tool binding involves providing the tool's name, description, and input schema to the LLM.
- Practically, this is done by creating a tool (e.g., a multiplication tool), then binding it to an LLM instance using a function like `bind_tools()`. The LLM with bound tools can then decide to call these tools when needed during interaction.

## Tool Calling: How LLMs Decide to Use Tools
- **Tool calling** is the process where the LLM decides during a conversation whether to use a specific tool based on the task. For example, if asked a simple question like "How are you?", the LLM will respond directly without calling any tool. But if asked "What is 8 * 7?", the LLM will generate a structured output specifying the tool to call (e.g., the multiplication tool) and the input arguments (A=8, B=7) instead of a plain text answer.
- Importantly, the LLM does **not** execute the tool itself; it only suggests which tool to use and with what inputs. The actual execution is handled by LangChain or the programmer. This separation is crucial for safety and control, preventing the LLM from autonomously running potentially risky operations.

## Tool Execution: Running the Suggested Tools
- After the LLM suggests a tool call, the programmer or LangChain executes the tool by passing the suggested inputs. This step is called **tool execution**. The tool runs with the provided inputs and returns the result. This result can then be wrapped in a special **tool message** and sent back to the LLM to inform it of the outcome.

## Maintaining Conversation Context with Messages
- The conversation history is maintained as a list of messages including human messages, AI messages, and tool messages. This history is sent back to the LLM to provide full context for generating the final answer. This message-based flow ensures the LLM has all necessary information from the conversation and tool executions to respond accurately.

## Real-Time Currency Conversion Application Example
- A practical example is built where two tools are created: one to fetch the real-time currency conversion factor by hitting an external API, and another to multiply the base currency value by the conversion factor. This solves the problem that LLMs cannot access real-time data on their own.
- These tools are then bound to the LLM, and queries like "What is the conversion factor between USD and INR?" and "Convert 10 USD to INR" are answered by the LLM calling the appropriate tools sequentially. This demonstrates how tool calling and execution can solve real-world problems with dynamic data.

## Handling Tool Argument Injection
- A challenge arises when one tool's output is needed as an input argument for another tool. To handle this, LangChain supports **injected tool arguments**, where the developer manually injects certain argument values after the first tool's execution, preventing the LLM from guessing or filling them incorrectly. This ensures correct sequential execution of dependent tools.

## Sequential Tool Execution Logic
- The code logic involves looping through tool calls suggested by the LLM, executing the first tool to get the conversion rate, injecting this rate into the second tool's arguments, and then executing the second tool to get the final converted value. The results of each tool execution are appended to the message history to maintain context.

## Difference Between Tool-Using Application and AI Agent
- The created application is **not an AI agent** because it is not autonomous. The programmer manually controls the flow and decision-making, such as when to call tools and how to handle outputs. A true AI agent autonomously breaks down problems and calls tools step-by-step without external intervention. This distinction is important for understanding the capabilities of LangChain-based applications.
